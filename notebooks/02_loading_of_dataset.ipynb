# 2. Loading of Dataset

In this section, we will be loading the Dataset provided while adhereing to the convention stipulated within the instructions provided for the assessment. We will be loading the Dataset into 3 different sets which are the Training Data, Testing Data and Validation Data. This is due to the data being pre-separated within the Dataset folder. ## 2.1 Loading & Data Pre-processing (23px by 23px)

In this sub-section, we will be loading in the Data for the 23px by 23px requirement from the Dataset Folder provided. All the Data Loaded will have the following attributes:

**Training Data (*train_generator_23*)**
- Normalised: **1/255**
- Target Size: **23 x 23**
- Batch Size: **32**
- Colour Mode: **Grayscale**
- Shuffle: **Shuffled**
- Augmentation: **Not Augmented**

**Validation Data (*val_generator_23*)**
- Normalised: **1/255**
- Target Size: **23 x 23**
- Batch Size: **32**
- Colour Mode: **Grayscale**
- Shuffle: **Not Shuffled**
- Augmentation: **Not Augmented**

**Testing Data (*test_generator_23*)**
- Normalised: **1/255**
- Target Size: **23 x 23**
- Batch Size: **32**
- Colour Mode: **Grayscale**
- Shuffle: **Not Shuffled**
- Augmentation: **Not Augmented**

**Augmented Training Data (*train_generator_23_aug*)**
- Normalised: **1/255**
- Target Size: **23 x 23**
- Batch Size: **32**
- Colour Mode: **Grayscale**
- Shuffle: **Shuffled**
- Augmentation: **Augmented**

With the attributes listed, we will proceed to Load and Process the Data in the Code Cell below.

```python
# ======================== Defining Dataset Paths ======================== #
base_dir = '/content/DELE_CA1/Dataset (A)'

train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

# ======================== Data Generators With Normalisation ======================== #
train_datagen_no_aug = ImageDataGenerator(rescale=1./255)
val_datagen_no_aug = ImageDataGenerator(rescale=1./255)
test_datagen_no_aug = ImageDataGenerator(rescale=1./255)
train_datagen_aug = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

# ======================== Loading of Training Data (23px by 23px) ======================== #
train_generator_23 = train_datagen_no_aug.flow_from_directory(
    train_dir,
    target_size=(23, 23),
    batch_size=32,
    color_mode='grayscale',
    class_mode='categorical',
    shuffle=True,
    seed=42
)

# ======================== Loading of Validation Data (23px by 23px) ======================== #
val_generator_23 = val_datagen_no_aug.flow_from_directory(
    val_dir,
    target_size=(23, 23),
    batch_size=32,
    color_mode='grayscale',
    class_mode='categorical',
    shuffle=False,
    seed=42
)


# ======================== Loading of Testing Data (23px by 23px) ======================== #
test_generator_23 = test_datagen_no_aug.flow_from_directory(
    test_dir,
    target_size=(23, 23),
    batch_size=32,
    color_mode='grayscale',
    class_mode='categorical',
    shuffle=False,
    seed=42
)

# ======================== Loading of Augmented Training Data (23px by 23px) ======================== #
train_generator_23_aug = train_datagen_aug.flow_from_directory(
    train_dir,
    target_size=(23, 23),
    batch_size=32,
    color_mode='grayscale',
    class_mode='categorical',
    shuffle=True,
    seed=42
)
```

With reference to the output of the Code Cell above, we are able to confirm that the data have been successfully loaded and we are able to proceed to load in the next set of data (101px by 101px) in accordance with the instructions provided.
