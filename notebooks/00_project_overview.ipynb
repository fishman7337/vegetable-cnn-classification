# Preliminary Background Context Reserch

**Cauliflower and Broccoli**
- Belong to the Brassicaceae family.
- Cauliflower has compact, white florets; broccoli has green, bumpy crowns.
- CNN-identifiable traits:
  - Fine-grained, clustered texture of the curd or crown.
  - Colour differences (white for cauliflower, green for broccoli).
  - Rounded dome shape with repetitive patterns.

**Bean**
- Includes various legumes like green beans and long beans.
- Characterised by thin, elongated, and slightly curved pods.
- CNN-identifiable traits:
  - Narrow, cylindrical shape.
  - Smooth surface with visible seams or tips.
  - Consistent greenish tones across samples.

**Radish and Carrot**
- Root vegetables from the Apiaceae family.
- Carrots are typically orange, while radishes are white or red.
- CNN-identifiable traits:
  - Long, tapering cylindrical shape.
  - High colour contrast with background (especially for orange carrots).
  - Surface texture and root hairs may be partially visible.

**Capsicum**
- Also known as bell peppers; belong to the Solanaceae family.
- Rounded and lobed with colour variation (green, red, yellow).
- CNN-identifiable traits:
  - Glossy, smooth surface with visible lobes.
  - Distinct shine and reflection patterns.
  - Bright, saturated colours that differ significantly from other classes.

**Cucumber and Bottle Gourd**
- Cucumber: elongated, smooth or slightly ridged skin.
- Bottle Gourd: bulbous lower portion with narrow upper neck.
- CNN-identifiable traits:
  - Shape profile is a key differentiator (straight vs bottle-like).
  - Surface texture and uniform pale green tones for gourd.
  - Fine differences in skin texture and curvature.

**Brinjal**
- Typically dark purple in colour, sometimes striped.
- Smooth, glossy surface with a green calyx at the top.
- CNN-identifiable traits:
  - High contrast between purple body and green crown.
  - Smooth reflection and curved shape.
  - Bulbous or elongated form with consistent hue.

**Pumpkin**
- Member of the Cucurbitaceae family.
- Large, round fruit with vertical ribbing and hard skin.
- CNN-identifiable traits:
  - Rounded shape with clear vertical grooves.
  - Matte skin surface with distinct colour (orange or dark green).
  - Visible stem and overall large size compared to others.

**Tomato**
- Belongs to the Solanaceae family.
- Round, smooth red fruit often attached to a green calyx.
- CNN-identifiable traits:
  - Uniform spherical shape with rich red colour.
  - Smooth, shiny surface with central brightness from lighting.
  - Occasional green stem contrast adds feature variation.

**Cabbage**
- Leafy vegetable with tightly packed leaves forming a spherical head.
- Light green or purple in colour with layered structure.
- CNN-identifiable traits:
  - Concentric leaf patterns and visible veins.
  - Pale, blended green/white tones.
  - Wrapping layers and rounded contour.

**Bitter Gourd**
- Known for its ridged, bumpy surface and elongated shape.
- Green in colour with pointed ends.
- CNN-identifiable traits:
  - Distinct wrinkled and textured skin.
  - Highly irregular surface geometry.
  - Strong edge contrast due to the raised ridges.

**Onion**
- Bulb-shaped with dry, flaky outer skin.
- Colours include white, red, or yellow.
- CNN-identifiable traits:
  - Symmetric, layered bulb shape.
  - Slight surface gloss and dry stem presence.
  - Rounded form and subtle shadow gradient.
